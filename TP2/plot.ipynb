```python
# De 100 a 5000, de 100 em 100, Bubble Sort
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo de resultados
resultado_file = "./tempo/ex1/bs.txt"

# Criar um DataFrame a partir do arquivo
df = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])

# Plotar o gráfico de linhas
plt.plot(df["Num"], df["Tempo"], marker='o')
plt.title("Bubble Sort - Tempo de execução por caso de teste")
plt.xlabel("Número de vértices")
plt.ylabel("Tempo de Execução (segundos)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# De 100 a 5000, de 100 em 100, Selection Sort
resultado_file_ss = "./tempo/ex1/ss.txt"
df_ss = pd.read_csv(resultado_file_ss, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, Bubble Sort
resultado_file_bs = "./tempo/ex1/bs.txt"
df_bs = pd.read_csv(resultado_file_bs, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, Insertion Sort
resultado_file = "./tempo/ex1/is.txt"
df_is = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, Quick Sort
resultado_file = "./tempo/ex1/qs.txt"
df_qs = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, MergeSort
resultado_file = "./tempo/ex1/ms.txt"
df_ms = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, HeapSort
resultado_file = "./tempo/ex1/hs.txt"
df_hs = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])

# De 100 a 5000, de 100 em 100, "Meu método"
resultado_file = "./tempo/ex1/eu.txt"
df_eu = pd.read_csv(resultado_file, sep=" ", header=None, names=["Num", "Tempo"])


# Criar uma figura e eixo compartilhado
fig, ax = plt.subplots()

# Plotar o gráfico do Bubble Sort
ax.plot(df_bs["Num"], df_bs["Tempo"], marker='o', label="Bubble Sort")

# Plotar o gráfico do Selection Sort
ax.plot(df_ss["Num"], df_ss["Tempo"], marker='o', color='red', label="Selection Sort")

#Plotar o gráfico do Insertion Sort
ax.plot(df_is["Num"], df_is["Tempo"], marker='o', color='green', label="Insertion Sort")

# Plotar o gráfico do Quick Sort
ax.plot(df_qs["Num"], df_qs["Tempo"], marker='o', color='black', label="Quick Sort")

# Plotar o gráfico do MergeSort
ax.plot(df_ms["Num"], df_ms["Tempo"], marker='o', color='purple', label="MergeSort")

# Plotar o gráfico do HeapSort
ax.plot(df_hs["Num"], df_hs["Tempo"], marker='o', color='orange', label="HeapSort")

# Plotar o gráfico do Meu método
ax.plot(df_eu["Num"], df_eu["Tempo"], marker='o', color=(0.87, 0.12, 0.57), label="Meu método")

# Adicionar legenda
ax.legend()

# Adicionar títulos e rótulos aos eixos
plt.title("Tempo de execução por caso de teste")
plt.xlabel("Número de vértices")
plt.ylabel("Tempo de Execução (segundos)")

# Adicionar grade
plt.grid(True)

# Exibir o gráfico
plt.show()
```
